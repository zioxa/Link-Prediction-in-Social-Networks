```python
%matplotlib inline

import random

import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
data = open('dataset 1.csv')
next(data)
g = nx.read_edgelist(data, create_using=nx.DiGraph(), delimiter=',', nodetype=int)
```

```python
g = nx.convert_node_labels_to_integers(G=g, first_label=0)
```

```python
sampled_edges = random.sample(list(g.edges), 500)
sampled_g = nx.DiGraph()
sampled_g.add_edges_from(sampled_edges)
sampled_g = nx.convert_node_labels_to_integers(G=sampled_g, first_label=0)
subg = nx.spring_layout(sampled_g)
nx.draw(sampled_g, subg, node_color='#1F77B4', edge_color='#02AE87', width=1, edge_cmap='YlGnBu', node_size=50)
```

```python
import pickle
import random
import time
st = time.time()

edges = dict()
for edge in g.edges:
    edges[(edge[0], edge[1])] = 1 # checking if edges are already present in the graph. If yes initialise to 1, else -1
    
missing_edges = set([])
count = 0

while (len(missing_edges)<g.number_of_edges()):
    if count % 50 == 0:
        print('iteration: ', count, '\tlen of miss edges: ', len(missing_edges), '\tedges left: ', 1768149-len(missing_edges))
    count += 1
    
    a=random.randint(0, 50000)
    b=random.randint(0, 50000)
    tmp = edges.get((a,b),-1) 
    if tmp == -1 and a!=b: #check if edge is present in the dictionary or not and that the node is not linked to itself
        try:
            # adding points who less likely to be friends
            if nx.shortest_path_length(g,source=a,target=b) > 2: 

                missing_edges.add((a,b))
            else:
                continue  
        except:  
                missing_edges.add((a,b))              
    else:
        continue
        
end = time.time()
print ('\n\ntime taken: ', end-st)
```

```python
f = open('missing_edges_final.p', 'wb')   # Pickle file is newly created 
pickle.dump(missing_edges, f) 
```
